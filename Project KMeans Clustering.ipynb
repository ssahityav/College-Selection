```python
tuition_cost = pd.read_csv("tuition_cost.csv")
salary_potential = pd.read_csv('salary_potential.csv')
salary_potential['average_salary'] = (salary_potential['early_career_pay'] + salary_potential['mid_career_pay'])/2
salary_potential
```

```python
salary_potential = salary_potential[['state_name', 'make_world_better_percent', 'stem_percent', 'average_salary']]
state_wise_salary = salary_potential.groupby(by=['state_name']).mean()
state_wise_salary = state_wise_salary.reset_index()
state_wise_salary
```

```python
tuition_cost = tuition_cost[['state','in_state_total', 'out_of_state_total']]
state_wise_tuition = tuition_cost.groupby(by=['state']).mean()
state_wise_tuition = state_wise_tuition.reset_index()
state_wise_tuition
```

```python
res = [sub.replace('-', ' ') for sub in state_wise_salary['state_name']]
state_wise_salary['state_name'] = res
state_wise_salary
```

```python
merged_table = state_wise_tuition.merge(state_wise_salary, how = 'inner', left_on = 'state', right_on = 'state_name')
merged_table = merged_table[['state', 'in_state_total', 'out_of_state_total', 'make_world_better_percent', 'stem_percent', 'average_salary']]
merged_table
```

## Diversity

```python

diversity_clean = diversity[(diversity['category'] == 'Black') | 
          (diversity['category'] == 'Hispanic') |
          (diversity['category'] == 'American Indian / Alaska Native') |
          (diversity['category'] == 'Asian') |
          (diversity['category'] == 'Native Hawaiian / Pacific Islander') |
          (diversity['category'] == 'White') |
          (diversity['category'] == 'Two Or More Races')]
```

```python
## Find fraction of total enrollment
enrollment_by_ethnicity = diversity_clean.groupby(by=['name'])['enrollment'].agg(['sum']).sort_values(by=['sum'],ascending=False)
enrollment_by_ethnicity.columns=['Total enrollment by ethnicity']
enrollment_by_ethnicity
```

```python
diversity_clean = diversity_clean.merge(enrollment_by_ethnicity, how='left',left_on='name',right_on='name')
diversity_clean
```

```python
## Clean the data

diversity_clean = diversity_clean[diversity_clean['Total enrollment by ethnicity'] >=354]
diversity_clean
```

```python
diversity_clean['Fraction of enrollment'] = diversity_clean['enrollment'] / diversity_clean['Total enrollment by ethnicity']
diversity_clean
```

```python
diversity_clean = diversity_clean[['name', 'state','category','enrollment','Total enrollment by ethnicity','Fraction of enrollment']]
```

```python
diversity_clean_max_fraction = diversity_clean.groupby('name')['Fraction of enrollment'].agg(['max'])
diversity_clean_max_fraction
```

```python
diversity_clean_max_fraction['max']>=0.75
diversity_clean_max_fraction['Diverse?'] = ((diversity_clean_max_fraction['max']<0.75).astype('int'))
diversity_clean_max_fraction = diversity_clean_max_fraction.reset_index()
```

```python
diversity_final = diversity_clean_max_fraction.merge(diversity_clean,left_on='name',right_on='name',how='inner')
diversity_final
diversity_final_mask = diversity_final
```

```python
diversity_final = diversity_final[['state','name','Diverse?']]
diversity_final = diversity_final.drop_duplicates()
diversity_final
```

```python
diversity_final = diversity_final.groupby('state')['Diverse?'].agg(['mean'])
diversity_final
```

```python
diversity_final = diversity_final.reset_index()
diversity_final
```

```python
diversity_final.columns=['state','fraction_of_diverse_colleges']
diversity_final
```

```python
merged_table = merged_table.merge(diversity_final,on='state')
merged_table
```

```python
import matplotlib.pyplot as plt
from kneed import KneeLocator
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
```

```python
features = merged_table.iloc[:,1:]
features
```

```python
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

```python
scaled_features[:,0] = scaled_features[:,0]*0.025
scaled_features[:,1] = scaled_features[:,1]*0.025
scaled_features[:,2] = scaled_features[:,2]*0.2375
scaled_features[:,3] = scaled_features[:,3]*0.2375
scaled_features[:,4] = scaled_features[:,4]*0.2375
scaled_features[:,5] = scaled_features[:,5]*0.2375
scaled_features
```

```python
n_classes = list(range(2,50))
SSE_elbow = [0]* len(list(range(2,50)))
i = 0

for n in n_classes:
        kmeans = KMeans(init="random",n_clusters=n,n_init=10,max_iter=300,random_state=42)
        kmeans.fit(scaled_features)
        SSE_elbow[i] = kmeans.inertia_
        
        i = i+1

SSE_elbow
```

```python
n_classes = list(range(2,50))
SSE = [0]* len(list(range(2,50)))
i = 0

for n in n_classes:
        kmeans = KMeans(init="random",n_clusters=n,n_init=10,max_iter=300,random_state=42)
        kmeans.fit(scaled_features)
        SSE[i] = kmeans.inertia_
        
        if np.bincount(kmeans.labels_).min() < 5:
            break
        
        i = i+1
        
SSE
```

```python
plt.style.use("fivethirtyeight")
plt.plot(list(range(2,50)), SSE_elbow)
plt.xticks(list(range(2,50)))
plt.xlabel("Number of Clusters")
plt.ylabel("SSE")
plt.show()
```

```python
kl = KneeLocator(list(range(2,50)), SSE_elbow, curve="convex", direction="decreasing")
kl.elbow
```
